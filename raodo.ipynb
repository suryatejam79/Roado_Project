Starting Hypothesis Testing H0 and H1

H0: Feature usage does not affect session duration
H1: Feature usage increases session duration

Using t-test, p-value = ___
Since p < 0.05 → reject H0

Conclusion: Feature usage leads to higher engagement

so, we have to find p value first on the basis of that we must choose to do that we are following below

```python
from scipy.stats import ttest_ind

feature_users = merged_df[merged_df['feature'].notnull()]['session_duration_sec']
non_feature_users = merged_df[merged_df['feature'].isnull()]['session_duration_sec']

t_stat, p_value = ttest_ind(feature_users, non_feature_users, equal_var=False)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

Test Used:
Independent Two-Sample t-test

Result:
p-value = 0.6199

Decision:
Since p-value > 0.05, we fail to reject the null hypothesis

Conclusion:
There is no statistically significant difference in session duration between users who use features and those who do not

```python
#What is engagement segmentation?
#We group users/customers based on how active they are
# and as per that I'm counting total session duration per user 
```

```python
#lets create engagment metric
engagement = merged_df.groupby('customer_id')['session_duration_sec'].sum().reset_index()

engagement.rename(columns={'session_duration_sec': 'total_session_time'}, inplace=True)

engagement.head()
```

```python
merged_df = merged_df.merge(engagement[['customer_id', 'segment']], on='customer_id', how='left')
```

```python
engagement['segment'] = pd.qcut(
    engagement['total_session_time'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
engagement['segment'].value_counts()
```

Segmentation Type:
Engagement-Based Segmentation

Methodology:
Customers were segmented based on total session duration. The total session time per customer was calculated and divided into three groups using quantile-based binning (Low, Medium, High).

Segments Created:

High Engagement (402 customers):
Users with the highest total session duration, indicating strong interaction with the platform
Medium Engagement (400 customers):
Users with moderate activity levels
Low Engagement (402 customers):
Users with minimal platform usage

Business Implications
High Engagement:
Likely loyal customers → target for retention and upselling
Medium Engagement:
Opportunity to increase engagement through targeted campaigns
Low Engagement:
At risk → require re-engagement strategies (offers, notifications, onboarding improvements)
